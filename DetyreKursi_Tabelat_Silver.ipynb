```python
import numpy as np
import pandas as pd
```

```python
teams = pd.read_csv("bronxe/teams.csv")
matches = pd.read_csv("bronxe/matches.csv")
players = pd.read_csv("bronxe/players.csv")
penalty_kicks = pd.read_csv("bronxe/penalty_kicks.csv")
goals = pd.read_csv("bronxe/goals.csv")
groups = pd.read_csv("bronxe/groups.csv")
group_standings = pd.read_csv("bronxe/group_standings.csv")
award_winners = pd.read_csv("bronxe/award_winners.csv")
player_appearances = pd.read_csv("bronxe/player_appearances.csv")
tournaments = pd.read_csv("bronxe/tournaments.csv")
tournament_standings = pd.read_csv("bronxe/tournament_standings.csv")
qualified_teams = pd.read_csv("bronxe/qualified_teams.csv")
squads = pd.read_csv("bronxe/squads.csv")
awards = pd.read_csv("bronxe/awards.csv")
```

```python
teams.dropna()
teams.drop_duplicates()

matches.dropna()
matches.drop_duplicates()

players.dropna()
players.drop_duplicates()

penalty_kicks.dropna()
penalty_kicks.drop_duplicates()

goals.dropna()
goals.drop_duplicates()

groups.dropna()
groups.drop_duplicates()

group_standings.dropna()
group_standings.drop_duplicates()

award_winners.dropna()
award_winners.drop_duplicates()

player_appearances.dropna()
player_appearances.drop_duplicates()

tournaments.dropna()
tournaments.drop_duplicates()

qualified_teams.dropna()
qualified_teams.drop_duplicates()

squads.dropna()
squads.drop_duplicates()

tournament_standings.dropna()
tournament_standings.drop_duplicates()


awards.dropna()
awards.drop_duplicates()
```

```python
teams.drop("federation_name", axis=1, inplace=True)
teams.drop("confederation_name", axis=1, inplace=True)
teams["team_name"] = teams["team_name"] + ", "+ teams["team_code"]
teams.drop("team_code", axis=1, inplace=True)
teams.drop("confederation_id", axis=1, inplace=True)
teams.drop("confederation_code", axis=1, inplace=True)
teams.drop("mens_team_wikipedia_link", axis=1, inplace=True)
teams.drop("womens_team_wikipedia_link", axis=1, inplace=True)
teams.drop("federation_wikipedia_link", axis=1, inplace=True)
teams.drop("key_id", axis=1, inplace=True)
teams=teams.loc[:, ~teams.columns.str.contains('^Unnamed')]
teams.to_csv("teams_new.csv")
teams.head()
```

```python
players.drop("player_wikipedia_link", axis=1, inplace=True)
players.drop("key_id", axis=1, inplace = True)
players=players.loc[:, ~players.columns.str.contains('^Unnamed')]
players.to_csv("players_new.csv")
players.head()
```

```python
matches.drop("replayed", axis=1, inplace=True)
matches.drop("replay", axis=1, inplace=True)
matches.drop("match_date", axis=1, inplace=True)
matches.drop("match_time", axis=1, inplace=True)
matches.drop("stadium_name", axis=1, inplace=True)
matches.drop("stadium_id", axis=1, inplace=True)
matches.drop("city_name", axis=1, inplace=True)
matches.drop("country_name", axis=1, inplace=True)
matches.drop("extra_time", axis=1, inplace=True)
matches.drop("penalty_shootout", axis=1, inplace=True)
matches.drop("score_penalties", axis=1, inplace=True)
matches.drop("home_team_score_penalties", axis=1, inplace=True)
matches.drop("away_team_score_penalties", axis=1, inplace=True)
matches.drop("group_name", axis=1, inplace=True)
matches.drop("key_id", axis=1, inplace=True)

matches["home_team_name"] = matches["home_team_name"] + ", " + matches["home_team_code"]
matches["away_team_name"] = matches["away_team_name"] + ", " + matches["away_team_code"]
matches=matches.loc[:, ~matches.columns.str.contains('^Unnamed')]
matches.to_csv("matches_new.csv")
matches.head()
```

```python
penalty_kicks.drop("tournament_id", axis=1, inplace=True)
penalty_kicks.drop("match_id", axis=1, inplace=True)
penalty_kicks.drop("match_date", axis=1, inplace=True)
penalty_kicks.drop("group_name", axis=1, inplace=True)
penalty_kicks.drop("team_code", axis=1, inplace=True)
penalty_kicks.drop("shirt_number", axis=1, inplace=True)
penalty_kicks.drop("stage_name", axis=1, inplace=True)
penalty_kicks.drop("team_id", axis=1, inplace=True)
penalty_kicks.drop("family_name", axis=1, inplace=True)
penalty_kicks.drop("given_name", axis=1, inplace=True)
penalty_kicks.drop("key_id", axis=1, inplace = True )
penalty_kicks=penalty_kicks.loc[:, ~penalty_kicks.columns.str.contains('^Unnamed')]
penalty_kicks.to_csv("penalty_kicks_new.csv")
penalty_kicks.head(10)
```

```python
goals = pd.read_csv("bronxe/goals.csv")

goals.drop("tournament_id", axis=1, inplace=True)
goals.drop("tournament_name", axis=1, inplace=True)
goals.drop("match_name", axis=1, inplace=True)
goals.drop("match_date", axis=1, inplace=True)

goals.drop("stage_name", axis=1, inplace=True)
goals.drop("group_name", axis=1, inplace=True)

goals.drop("team_id", axis=1, inplace=True)
goals.drop("team_name", axis=1, inplace=True)

goals.drop("team_code", axis=1, inplace=True)
goals.drop("home_team", axis=1, inplace=True)

goals.drop("away_team", axis=1, inplace=True)
goals.drop("family_name", axis=1, inplace=True)

goals.drop("given_name", axis=1, inplace=True)
goals.drop("shirt_number", axis=1, inplace=True)

goals.drop("player_team_name", axis=1, inplace=True)
goals.drop("player_team_code", axis=1, inplace=True)

goals.drop("minute_label", axis=1, inplace=True)
goals.drop("minute_regulation", axis=1, inplace=True)

goals.drop("minute_stoppage", axis=1, inplace=True)
goals.drop("match_period", axis=1, inplace=True)

goals.drop("own_goal", axis=1, inplace=True)
goals.drop("key_id", axis=1, inplace=True)
goals=goals.loc[:, ~goals.columns.str.contains('^Unnamed')]
goals.to_csv("goals_new.csv")
goals.head()
```

```python
groups.drop("tournament_name", axis=1, inplace=True)
groups=groups.loc[:, ~groups.columns.str.contains('^Unnamed')]
groups.to_csv("groups_new.csv")
groups.head()
```

```python
award_winners.drop("team_code",axis=1, inplace=True)
award_winners.drop("tournament_name",axis=1, inplace=True)
award_winners.drop("family_name",axis=1, inplace=True)
award_winners.drop("given_name",axis=1, inplace=True)
award_winners=award_winners.loc[:, ~award_winners.columns.str.contains('^Unnamed')]
award_winners.to_csv("award_winners_new.csv")
award_winners.head()
```

```python
player_appearances.drop("tournament_name",axis=1, inplace=True)
player_appearances.drop("match_date",axis=1, inplace=True)
player_appearances.drop("team_name",axis=1, inplace=True)
player_appearances.drop("team_code",axis=1, inplace=True)
player_appearances.drop("away_team",axis=1, inplace=True)
player_appearances.drop("family_name",axis=1, inplace=True)
player_appearances.drop("given_name",axis=1, inplace=True)
player_appearances.drop("substitute",axis=1, inplace=True)
player_appearances=player_appearances.loc[:, ~player_appearances.columns.str.contains('^Unnamed')]
player_appearances.to_csv("player_appearances_new.csv")
player_appearances.head()
```

```python
squads.drop("tournament_name",axis=1, inplace=True)
squads.drop("team_name",axis=1, inplace=True)
squads.drop("team_code",axis=1, inplace=True)
squads.drop("family_name",axis=1, inplace=True)
squads.drop("given_name",axis=1, inplace=True)
squads.drop("position_code",axis=1, inplace=True)
squads=squads.loc[:, ~squads.columns.str.contains('^Unnamed')]
squads.to_csv("squads_new.csv")
squads.head()
```

```python
group_standings.drop("tournament_name", axis=1, inplace=True)
group_standings.drop("team_name", axis=1, inplace=True)
group_standings.drop("team_code", axis=1, inplace=True)
group_standings.drop("goal_difference", axis=1, inplace=True)
group_standings=group_standings.loc[:, ~group_standings.columns.str.contains('^Unnamed')]
group_standings.to_csv("group_standings_new.csv")
group_standings.head()
```

```python
tournament_standings.drop("tournament_name", axis=1, inplace=True)
tournament_standings.drop("team_name", axis=1, inplace=True)
tournament_standings.drop("team_code", axis=1, inplace=True)
tournament_standings=tournament_standings.loc[:, ~tournament_standings.columns.str.contains('^Unnamed')]
tournament_standings.to_csv("tournament_standings_new.csv")
tournament_standings.head()
```
